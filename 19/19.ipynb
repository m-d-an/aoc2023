```python
file = open('input.txt', 'r')
input = file.readlines()
file.close()
```

```python
workflows = {}
breakline = 0
for i,line in enumerate(input):
    if line.strip() == '':
        breakline = i
        break
    else:
        linesplit = line.strip().split('{')
        name = linesplit[0]
        workflows[name] = []
        steps = linesplit[1][:-1].split(',')
        for step in steps:
            if ':' in step:
                stepsplit = step.split(':')
                workflows[name].append((stepsplit[0][0],stepsplit[0][1],int(stepsplit[0][2:]),stepsplit[1]))
            else:
                workflows[name].append(step)
#print(workflows)
```

```python
sum_values = 0
for line in input[i+1:]:
    values = {}
    for v in line[1:-2].split(','):
        vsplit = v.split('=')
        values[vsplit[0]] = int(vsplit[1])
    
    #go through workflows
    wf = 'in'
    while(wf != 'A' and wf != 'R'):
        for w in workflows[wf]:
            if type(w) == str:
                wf = w
                break
            else:
                if w[1] == '>':
                    if values[w[0]] > w[2]:
                        wf = w[3]
                        break
                    else:
                        continue
                else:
                    if values[w[0]] < w[2]:
                        wf = w[3]
                        break
                    else:
                        continue

    if wf == 'A':
        print(f"Accepted: {sum(values.values())}")
        sum_values += sum(values.values())
    else:
        print('Rejected')
print()
print(f"Result: {sum_values}")
```

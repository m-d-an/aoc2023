```python
file = open('input.txt', 'r')
input = file.readlines()
file.close()
```

```python
rows = len(input)
cols = len(input[0].strip())
rocks = []
start = (0,0)
reachable = []
```

```python
def print_grid():
    for i in range(rows):
        line = ""
        for j in range(cols):
            if (i,j) in rocks:
                line += '#'
            elif (i,j) in reachable:
                line += 'O'
            elif start == (i,j):
                line += 'S'
            else:
                line += '.'
        print(line)
    print()
```

```python

for i,row in enumerate(input):
    for j,char in enumerate(row.strip()):
        if char == '#':
            rocks.append((i,j))
        elif char == 'S':
            start = (i,j)
print_grid()
```

```python
last_steps = []
checked = []
steps = 0
last_steps.append(start)
while steps<64:
    steps += 1

    new_steps = []
    for coord in last_steps:
        #up
        if coord[0] > 0 and (coord[0]-1,coord[1]) not in rocks:
            new_steps.append((coord[0]-1,coord[1]))
        #down
        if coord[0] < rows-1 and (coord[0]+1,coord[1]) not in rocks:
            new_steps.append((coord[0]+1,coord[1]))
        #left
        if coord[1] > 0 and (coord[0],coord[1]-1) not in rocks:
            new_steps.append((coord[0],coord[1]-1))
        #right
        if coord[1] < cols-1 and (coord[0],coord[1]+1) not in rocks:
            new_steps.append((coord[0],coord[1]+1))
    #remove duplicates and already checked coords
    new_steps = list(set(new_steps) - set(checked))
    #print(f"new steps: {new_steps}")
    
    if steps%2==0:     
        reachable = reachable+new_steps
        #print(f"reachable: {reachable}")

    checked = checked + new_steps
    last_steps = new_steps

    
    #print_grid()
len(reachable)
```
